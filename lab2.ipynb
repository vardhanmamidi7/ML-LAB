```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate random data for X1 and X2
np.random.seed(42)
X1 = np.random.rand(100) * 10  # Random values between 0 and 10
X2 = np.random.rand(100) * 10  

# Define the true relationship: Y = 2*X1 + 3*X2 + 5 (with some noise)
Y = 2 * X1 + 3 * X2 + 5 + np.random.randn(100) * 2

# Reshape X1 and X2 into a feature matrix
X = np.column_stack((X1, X2))

# Train the multiple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Get the predicted plane
X1_grid, X2_grid = np.meshgrid(np.linspace(0, 10, 10), np.linspace(0, 10, 10))
Y_grid = model.intercept_ + model.coef_[0] * X1_grid + model.coef_[1] * X2_grid

# Plot the data and the regression plane
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(X1, X2, Y, color='blue', label='Data Points')

# Plot the regression plane
ax.plot_surface(X1_grid, X2_grid, Y_grid, color='red', alpha=0.5)

# Labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('3D Plane Separator in Multiple Linear Regression')

plt.legend()
plt.show()

```

```python
# 12. Write a function to find min and max values of any
# column

def find_min_max(df, column_name):
    """
    Function to find the minimum and maximum values of a given column.
    
    Parameters:
    df (pd.DataFrame): The dataset
    column_name (str): The column for which to find min and max values
    
    Returns:
    tuple: (min_value, max_value)
    """
    if column_name in df.columns:
        min_value = df[column_name].min()
        max_value = df[column_name].max()
        return min_value, max_value
    else:
        return f"Column '{column_name}' not found in dataset"

# Example usage:
horsepower_min_max = find_min_max(df, "Horsepower")
length_min_max = find_min_max(df, "Length")
fuel_efficiency_min_max = find_min_max(df, "Fuel efficiency")

horsepower_min_max, length_min_max, fuel_efficiency_min_max
```
