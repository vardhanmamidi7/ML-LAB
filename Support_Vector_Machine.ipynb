```python
# Assuming the features are all columns except 'Species' (which is our target)
X = iris.drop('Species', axis=1)
y = iris['Species']

# Check shapes
print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data (70% train, 30% test), using random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
# help(SVC)
```

```python
# Import SVC from sklearn.svm
from sklearn.svm import SVC

# Q23: Instantiate SVC with C=1 (default kernel is 'rbf' unless specified otherwise)
svm_model = SVC(C=1, random_state=1)
```

```python
# Q24: Fit the model using the training data
svm_model.fit(X_train, y_train)
```

```python
# Q25: Predict on the test set
y_pred = svm_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

# Q26: Print the confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_mat)

```

```python
sns.heatmap(conf_mat,annot=True,cmap="Blues")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Q27: Evaluate with accuracy and classification report
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score, KFold

# Q28: Perform K-Fold CV with 10 folds (for example)
kf = KFold(n_splits=10, shuffle=True, random_state=1)
cv_scores = cross_val_score(svm_model, X_train, y_train, cv=kf)

print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# Q30 & Q31: Define the parameter grid; try C values [0.1, 1, 10, 100, 1000]
param_grid = {'C': [0.1, 1, 10, 100, 1000]}

# Q32: Specify number of folds (e.g., use 10-fold CV)
grid_search = GridSearchCV(SVC(random_state=1), param_grid, cv=10, scoring='accuracy')

# Q33: Fit grid search on the training data
grid_search.fit(X_train, y_train)
```

```python
import pandas as pd

# Extract the grid search results and convert to a DataFrame for easier viewing
results_df = pd.DataFrame(grid_search.cv_results_)
print(results_df[['params', 'mean_test_score', 'std_test_score']])
```

```python
import matplotlib.pyplot as plt

# Extract mean scores for plotting
C_values = [result['C'] for result in results_df['params']]
mean_test_scores = results_df['mean_test_score']

plt.figure(figsize=(8, 6))
plt.plot(C_values, mean_test_scores, marker='o', linestyle='-', color='b')
plt.xscale('log')
plt.xlabel('C (Cost Parameter)')
plt.ylabel('Mean CV Accuracy')
plt.title('Grid Search CV: C vs. Accuracy')
plt.grid(True)
plt.show()
```

```python
# Q36: Display the best score and corresponding C value
print("Best CV Score:", grid_search.best_score_)
print("Best C value:", grid_search.best_params_['C'])
```
