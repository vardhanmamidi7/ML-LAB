# Difference between one hot encoder  the  pd .get_dummies()

When using pd.get_dummies(), it automatically handles everything within a single step, unlike OneHotEncoder, where we manually drop and re-add columns.

What Happens with pd.get_dummies()?
It creates new columns for each category in a categorical variable.
It replaces the original categorical column directly in the DataFrame.
If drop_first=True, it removes the first category to avoid the dummy variable trap.
It returns a new DataFrame with transformed categorical columns and keeps all other columns unchanged.
